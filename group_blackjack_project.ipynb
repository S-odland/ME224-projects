```
row = 57

from google.colab import auth
from oauth2client.client import GoogleCredentials
import gspread
import random
import time

auth.authenticate_user()
gc = gspread.authorize(GoogleCredentials.get_application_default())

myspreadsheet = gc.open_by_url('https://docs.google.com/spreadsheets/d/1QPVPPv37nzDVqSfnNkLXxYWJlmBwPfhfoK46bySGLx0/edit#gid=0')
blapblap = myspreadsheet.worksheet('blapblap')

# The name of the game is blackjack, well kind of
 
selectedcards = [1]

def randcard():

  y = 1
  while(y):
    card = random.randint(4,55)
    for i in range(len(selectedcards)):
      if card == selectedcards[i]:
        break
    else:
      selectedcards.append(card)
      y = 0

  return card
 
def cardinsheet(card,i,suit_col,val_col):

  card_suit = blapblap.cell(card,suit_col).value
  card_val = blapblap.cell(card,val_col).value

  blapblap.update_cell(58 + i,suit_col,card_suit)
  blapblap.update_cell(58 + i,val_col,card_val)

  return card_suit,card_val
  
def clearhand(i,suit_col,val_col):

  for i in range(i + 1):
    blapblap.update_cell(57 + i,suit_col,'')
    blapblap.update_cell(57 + i,val_col,'')

def convert(row,col,score):

  if blapblap.cell(row,col).value == 'J' or blapblap.cell(row,col).value == 'Q' or blapblap.cell(row,col).value == 'K':
    score += 10
  elif blapblap.cell(row,col).value == 'A':
    score += 11
  else:
    score += int(blapblap.cell(row,col).value)
  
  return score

blapblap.update_cell(1,5,0)

def play():
  row = 58
  row3 = row
  row2 = row
  row1 = row
  while 1: 
    selectedcards = [1]
    clearhand(row3 - 57,1,2)
    clearhand(row1 - 57,4,5)
    clearhand(row2 - 57,7,8)
    row = 58
    startrow = 58
    score1 = 0
    score2 = 0
    score3 = 0
    j = 0
    k = 0
    nos = 0

    ## now we got a dealer daemon and player daemons set with specific strategys
    ## the play atuomically, first scott's daemon will go, then alec's then JD's
    ## the score is settled at the end (after 3 while loops)

    for i in range(2):

      blapblap.update_cell(57,4,'Cards in Deck:')
      blapblap.update_cell(57,5,52 -(j+1))
      cardinsheet(randcard(),row - startrow,4,5)
      score1 = convert(row,5,score1)
      j += 1

      blapblap.update_cell(57,4,'Cards in Deck:')
      blapblap.update_cell(57,5,52 -(j+1))
      cardinsheet(randcard(),row - startrow,7,8)
      score2 = convert(row,8,score2)
      j += 1

      blapblap.update_cell(57,4,'Cards in Deck:')
      blapblap.update_cell(57,5,52 -(j+1))
      cardinsheet(randcard(),row - startrow,1,2)      
      score3 = convert(row,2,score3)
      j += 1

      row += 1

    blapblap.update_cell(row,4,'SCORE')
    blapblap.update_cell(row,5,score1)

    blapblap.update_cell(row,7,'SCORE')
    blapblap.update_cell(row,8,score2)

    blapblap.update_cell(row,1,'SCORE')
    blapblap.update_cell(row,2,score3)
    row += 1

    if score1 == 21:
      blapblap.update_cell(row,5,'Winner Winner Chicken Dinner')
      blapblap.update_cell(1,5,int(blapblap.cell(1,5).value) + 1)
      continue
    elif score2 == 21:
      blapblap.update_cell(row,8,'Winner Winner Chicken Dinner')
      blapblap.update_cell(1,8,int(blapblap.cell(1,8).value) + 1)
      continue
    elif score3 == 21:
      blapblap.update_cell(row,2,'Winner Winner Chicken Dinner')
      blapblap.update_cell(1,2,int(blapblap.cell(1,2).value) + 1)
      continue

    k = 2
    row1 = row
    row2 = row
    row3 = row

    while score1 < 21:

      blapblap.update_cell(row1,4,'Hit(y/n)?')
      if score1 < 17:
        blapblap.update_cell(row1,5,'y')
      else:
        blapblap.update_cell(row1,5,'n')

      row1 += 1

      if blapblap.cell(row1-1,5).value == 'y':

        cardinsheet(randcard(),row1 - startrow,4,5)
        k += 1
        # tells you how many card are left in the deck
        blapblap.update_cell(57,4,'Cards in Deck:')
        blapblap.update_cell(57,5,52 -k)
        
        score1 = convert(row1,5,score1)
        row1 += 1
      else:
        row1 += 1
        break
        
      blapblap.update_cell(row1,4,'SCORE')
      blapblap.update_cell(row1,5,score1)

      row1 += 1

      if score1 == 21:
        break

    while score2 < 21:
      
      blapblap.update_cell(row2,7,'Hit(y/n)?')
      if score2 < 13:
        blapblap.update_cell(row2,8,'y')
      else:
        blapblap.update_cell(row2,8,'n')
      row2 += 1

      if blapblap.cell(row2-1,8).value == 'y':

        cardinsheet(randcard(),row2 - startrow,7,8)
        k += 1
        # tells you how many card are left in the deck
        blapblap.update_cell(57,4,'Cards in Deck:')
        blapblap.update_cell(57,5,52 -k)
        
        score2 = convert(row2,8,score2)
        row2 += 1
      else:
        row2 += 1
        break
      
      blapblap.update_cell(row2,7,'SCORE')
      blapblap.update_cell(row2,8,score2)
      row2 += 1

      if score2 == 21:
        break

    while score3 < 21:

      blapblap.update_cell(row3,1,'Hit(y/n)?')
      if score3 < 16:
        blapblap.update_cell(row3,2,'y')
      else:
        blapblap.update_cell(row3,2,'n')
      row3 += 1

      if blapblap.cell(row3-1,2).value == 'y':

        cardinsheet(randcard(),row3 - startrow,1,2)
        k += 1
        # tells you how many card are left in the deck
        blapblap.update_cell(57,4,'Cards in Deck:')
        blapblap.update_cell(57,5,52 -k)
        
        score3 = convert(row3,2,score3)

        row3 += 1

      else:
        row3 += 1
        break
      
      blapblap.update_cell(row3,1,'SCORE')
      blapblap.update_cell(row3,2,score3)

      row3 += 1

      if score3 == 21:
        break

# below checks winners: winners are if you get 21, no ties, if none bust it checks greatest score for winner
# if one busts there are no winners only loooooosers

    if score1 < 21:
      blapblap.update_cell(row1,4,'FINAL SCORE')
      blapblap.update_cell(row1,5,score1)
      if score2 > 21 and score3 > 21:
        blapblap.update_cell(1,5,int(blapblap.cell(1,5).value) + 1)
      elif score2 < 21 and score3 < 21:
        if score1 > score2 and score1 > score3:
          blapblap.update_cell(1,5,int(blapblap.cell(1,5).value) + 1)
    elif score1 > 21:
      blapblap.update_cell(row1,4,'You are a')
      blapblap.update_cell(row1,5,'Busted Looser')
    elif score1 == 21:
      blapblap.update_cell(row1,5,'Winner Winner Chicken Dinner')
      # wins are shown at the top of teh sheet
      blapblap.update_cell(1,5,int(blapblap.cell(1,5).value) + 1)

    if score2 < 21:
      blapblap.update_cell(row2,7,'FINAL SCORE')
      blapblap.update_cell(row2,8,score2)
      if score1 > 21 and score3 > 21:
        blapblap.update_cell(1,8,int(blapblap.cell(1,8).value) + 1)
      elif score1 < 21 and score3 < 21:
        if score2 > score1 and score2 > score3:
          blapblap.update_cell(1,5,int(blapblap.cell(1,5).value) + 1)
    elif score2 > 21:
      blapblap.update_cell(row2,7,'You are a')
      blapblap.update_cell(row2,8,'Busted Looser')
    elif score2 == 21:
      blapblap.update_cell(row2,8,'Winner Winner Chicken Dinner')
      # wins are shown at the top of teh sheet
      blapblap.update_cell(1,8,int(blapblap.cell(1,8).value) + 1)

    if score3 < 21:
      blapblap.update_cell(row3,1,'FINAL SCORE')
      blapblap.update_cell(row3,2,score3)
      if score1 > 21 and score2 > 21:
        blapblap.update_cell(1,8,int(blapblap.cell(1,8).value) + 1)
      elif score2 < 21 and score1 < 21:
        if score3 > score2 and score3 > score1:
          blapblap.update_cell(1,5,int(blapblap.cell(1,5).value) + 1)
    elif score3 > 21:
      blapblap.update_cell(row3,1,'You are a')
      blapblap.update_cell(row3,2,'Busted Looser')
    elif score3 == 21:
      blapblap.update_cell(row3,2,'Winner Winner Chicken Dinner')
      # wins are shown at the top of teh sheet
      blapblap.update_cell(1,2,int(blapblap.cell(1,2).value) + 1)
       

play()
```
